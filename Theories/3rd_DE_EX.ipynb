## What is developing inside a container?

- Rather than working locally, can make your project work in VSCode through the Dev Container extension.
- It's recommended to keep the themes, UI extension of VSCode on local machine, and Workplace Extension for the Container.

    

## Getting started

- Add new Dev Config file &rarr; .devcontainer.json
    - name: project name
    - dockerComposeFile: path/to/docker-compose.yaml
    - service: name of the service to work with
    - workspaceFolder: path/to/open/vscode
- Open the container through the Command Palette.
- Wait for the container to build up and you're good to go.
