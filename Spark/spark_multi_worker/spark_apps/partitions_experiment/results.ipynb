# Execution time results for each setting of 12 cores using Spark Standalone Client mode

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 2 workers/cluster, 2 executors/worker, 3 cores/executor

```python
file_name = 'Exe_time_2_2_3'
execution_time = []
with open(f'./data/{file_name}.txt', 'r') as f:
       for line in f.readlines():
              execution_time.append(round(float(line), 3))

num_of_repartitions = range(4, 61, 4)
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(num_of_repartitions, execution_time)

ax.set(xlabel='Number of partitions', ylabel='Execution time (s)',
       title='Execution time on different repartitions \n with 2 workers/cluster, 2 exes/worker, 3 cores/exe')

x_ticks = np.arange(0, 65, 4)
y_ticks = np.arange(16, 26.1, 0.5)

ax.set_xticks(x_ticks)
ax.set_yticks(y_ticks, minor=True)
ax.grid()

fig.savefig(f'./plots/{file_name}.png')
plt.show()
```
