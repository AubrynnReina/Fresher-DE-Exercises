# 1. Indexing in MySQL

- Stored in B-trees (default)
- Except for spatial data types: uses R-trees
- Memory table has support for hash indexes

# 2. Demonstration

## 2.1. Data and Query

- Data: [Massive Bank dataset (1004480 rows)](https://www.kaggle.com/datasets/ksabishek/massive-bank-dataset-1-million-rows)

- Query:
```sql
    SELECT * FROM bankdataset
    WHERE Transaction_count IN (2003, 2000, 1213, 1789, 1555)
```

## 2.2. Result

### 2.2.1. Time-based

| Trials/Aggregate | No Index | With Index |
| --- | --- | --- |
| 1 | 7.9ms | 7.1ms |
| 2 | 5.2ms | 0.4ms |
| 3 | 4.6ms | 0.5ms |
| 4 | 5.2ms | 0.7ms |
| 5 | 5.3ms | 0.4ms |
| Mean (with init) | 5.64ms | 0.91ms |
| Mean (without init) | 5.075ms | 0.5ms |

&rarr; Indexing produces result almost 10 times faster than not having indexed the `Transaction_count` column
